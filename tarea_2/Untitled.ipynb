```python
def min_dist(x1,x2,y1,y2):
    return np.sqrt((x2-x1)**2 + (y2-y1)**2)
```

```python
fig=plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
for i in range(len(P[:,0])):
    plt.plot(P[i,0], P[i,1], marker='o', c=c[i],markersize=8, linestyle = 'None', label="Punto "+str(i+1))
plt.plot([P[n//2,0]]*10, np.arange(-1,9,1),c='k')
plt.xticks(np.arange(min(P[:,0])-1, max(P[:,0])+1, 1), fontsize=14)
plt.yticks(np.arange(min(P[:,1])-1, max(P[:,1])+1, 1), fontsize=14)
plt.annotate(s='', xy=(P[1,0],P[1,1]), xytext=(P[2,0],P[2,1]), arrowprops=dict(arrowstyle='<->'))
rect = patches.Rectangle((4,8),min_dist(P[1,0],P[2,0],P[1,1],P[2,1]),-8,linewidth=1,edgecolor='r',alpha=0.3)
rect2 = patches.Rectangle((4-min_dist(P[1,0],P[2,0],P[1,1],P[2,1]),8),min_dist(P[1,0],P[2,0],P[1,1],P[2,1]),-8,linewidth=1,edgecolor='r',alpha=0.3)
ax.add_patch(rect)
ax.add_patch(rect2)
plt.annotate(s='', xy=(4-min_dist(P[1,0],P[2,0],P[1,1],P[2,1]),7), xytext=(4+min_dist(P[1,0],P[2,0],P[1,1],P[2,1]),7), arrowprops=dict(arrowstyle='<->'))
plt.text(3.1,4.5,"dl",fontsize=15)
plt.text(3.5,6.7,"d",fontsize=15)
plt.text(4.4,6.7,"d",fontsize=15)
plt.xlim(min(P[:,0]-.1),max(P[:,0]+.1))
plt.ylim(min(P[:,1]-.1),max(P[:,1]+.1))
plt.grid()
plt.legend(loc="upper left")
#plt.show()
plt.savefig("figs/f5.pdf",bbox_inches = "tight", dpi=300)
```

```python
def create_file(N,M,s,number):
    np.random.seed(s)
    f = open("test_1/input" + str(number) + ".txt", "a")
    for i in range(M):
        Px = []
        Py = []
        f.write(str(N)+"\n")
        for j in range(N):
            x,y = np.random.uniform(0, 1000), np.random.uniform(0, 1000)
            if x not in Px and y not in Py:
                f.write(str(round(x,3)) + " " + str(round(y,3))+"\n")
            Px.append(x)
            Py.append(y)
        
    f.close()
```

```python
def swap_subarray(a, l, r, m):
    for i in range(m):
        a[l+i],a[r-i] = a[r-i],a[l+i]
    return a
```

```python
 swap_subarray(a, 300, n, 5)
```

```python
def create_file_2(a, l, r, m, number):
    f = open("test_2/input" + str(number) + ".txt", "a")
    a = swap_subarray(a, l, r, m)
    for i in range(len(a)):
        f.write(str(a[i])+"\n")
    f.close()
```
